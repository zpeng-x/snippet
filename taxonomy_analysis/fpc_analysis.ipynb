```python
import pandas as pd
```

```python
# Load taxonomy.en-US.txt file
taxonomy_file = 'taxonomy.en-US.txt'
taxonomy_df = pd.read_csv(taxonomy_file, sep='\t', header=None, names=['Category'])
print(taxonomy_df.head())

text_data = taxonomy_df['Category'].str.cat(sep='\n')  
```

```python
# Alternative: Read from a file
# with open('taxonomy.txt', 'r') as file:
#     text_data = file.read()

# Convert text data to a list of lines, stripping whitespace
lines = [line.strip() for line in text_data.strip().split('\n') if line.strip()]

# Split each line on '>' and create a list of taxonomy levels
taxonomy_data = [line.split(' > ') for line in lines]

# Find the maximum number of levels to determine the number of columns
max_levels = max(len(taxonomy) for taxonomy in taxonomy_data)

# Create column names dynamically (e.g., Level_1, Level_2, ...)
columns = [f'Level_{i+1}' for i in range(max_levels)]

# Create a DataFrame by padding shorter lists with None
df = pd.DataFrame(
    [taxonomy + [None] * (max_levels - len(taxonomy)) for taxonomy in taxonomy_data],
    columns=columns
)

# Display the DataFrame
print(df)

# Optional: Save to CSV for further use
df.to_csv('taxonomy_output.csv', index=False)
```

```python
# Read facebook 
fb = pd.read_csv('fb_product_categories_en_US.csv')
```

```python
# Parse taxonomy column on category
fb_taxonomy_data = [cat.split(' > ') for cat in fb['category']]
```

```python
# Organize into DataFrame
max_fb_levels = max(len(taxonomy) for taxonomy in fb_taxonomy_data)
fb_columns = [f'Level_{i+1}' for i in range(max_fb_levels)]

fb_df = pd.DataFrame(
    [taxonomy + [None] * (max_fb_levels - len(taxonomy)) for taxonomy in fb_taxonomy_data],
    columns=fb_columns
)
# Save fb_df
fb_df.to_csv("fb_taxonomy_output.csv", index=False)
```
